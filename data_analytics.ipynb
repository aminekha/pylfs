```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
# paths
data_dir = "data/"
images_dir = os.path.join(data_dir, "IMG")
```

```python
# names = ["image", "forward", "left", "right", "stop"]
main_df = pd.read_csv(os.path.join(data_dir, "driving_log.csv"))
```

```python
print("Number of lefts")
main_df['left'].value_counts()
```

```python
print("Number of Right")
main_df['right'].value_counts()
```

```python
print("Number of stops")
main_df['stop'].value_counts()
```

```python
# number of zeroes and number of ones
# main_df[main_df[["forward", "left", "right", "stop"]] == 0].value_counts()
zeroes_forward = main_df[main_df["forward"] == 0].shape[0]
ones_forward = main_df[main_df["forward"] == 1].shape[0]
zeroes_left = main_df[main_df["left"] == 0].shape[0]
ones_left = main_df[main_df["left"] == 1].shape[0]
zeroes_right = main_df[main_df["right"] == 0].shape[0]
ones_right = main_df[main_df["right"] == 1].shape[0]
zeroes_stop = main_df[main_df["stop"] == 0].shape[0]
ones_stop = main_df[main_df["stop"] == 1].shape[0]

print(zeroes_forward)
print(ones_forward)
print()
print(zeroes_left)
print(ones_left)
print()
print(zeroes_right)
print(ones_right)
print()
print(zeroes_stop)
print(ones_stop)
```

```python
# visualize the data

forward_zeroes, forward_ones = main_df[main_df["forward"] == 0], main_df[main_df["forward"] == 1]
left_zeroes, left_ones = main_df[main_df["left"] == 0], main_df[main_df["left"] == 1]

ind = np.arange(len(forward_zeroes))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, forward_zeroes, width, yerr=forward_ones,
                label='Men')
rects2 = ax.bar(ind + width/2, left_zeroes, width, yerr=left_ones,
                label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()

plt.show()
```
